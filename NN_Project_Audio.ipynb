```python
# Extracting MFCC's For each and every audio file
import pandas as pd
import os
import librosa

path_dataset='Audio_dataset\\UrbanSound8K\\audio\\'
```

```python
# Defination to extract MFCC's for every audio file
def mfccs_extractor(file):
    audio, sample_rate = librosa.load(file, res_type='kaiser_fast') 
    mfccs_features = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=50)
    mfccs_scaled_features = np.mean(mfccs_features.T,axis=0)
    
    return mfccs_scaled_features
```

```python
# Iterating through every audio file and extract MFCC to extract the feature
# MFCCs(Mel-Frequency Cepstral Coefficients)
import numpy as np
from tqdm import tqdm

extracted_features=[]
for index,row in tqdm(metadata.iterrows()):
    file = os.path.join(os.path.abspath(path_dataset),'fold'+str(row["fold"])+'/',str(row["slice_file_name"]))
    class_labels=row["class"]
    data=mfccs_extractor(file)
    extracted_features.append([data,class_labels])
```

```python
# converting extracted_features to Pandas dataframe
extracted_feature_data=pd.DataFrame(extracted_features,columns=['feature','class'])
extracted_feature_data.head(5)
```

```python
# Split the dataset to get independent dataset
X=np.array(extracted_feature_data['feature'].tolist())

print(f'Shape of independent variable : {X.shape}')
print(f'\n Independent Variable values: \n {X}')
```

```python
# Split the dataset to get dependent dataset
y=np.array(extracted_feature_data['class'].tolist())
print(f'Shape of dependent variable : {y.shape}')
print(f'\n Dependent Variable values: \n {y}')
```

```python
# one hot encoding
y_one_hot = np.array(pd.get_dummies(y))

print(f'Shape of Encoded y : {y_one_hot.shape}')
print(f'Encoded y :\n {y_one_hot}')
```

## Random Forest

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Create Random Forest classifier
rf_clf = RandomForestClassifier(n_estimators=100)

# Train the classifier
rf_clf.fit(X_train, y_train)

# Predict on the test set
predict_rf = rf_clf.predict(X_test)
```

```python
# Evaluate accuracy
accuracy_rf = accuracy_score(y_test, predict_rf) * 100
print(f'Test Accuracy of Random Forest: {accuracy_rf} %')
```

```python
from sklearn.metrics import confusion_matrix
rf_cnm = confusion_matrix(y_test, predict_rf)
print(f'Confusion Matrix of Random Forest : \n{rf_cnm}')
```

## ANN

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y_one_hot,test_size=0.2,random_state=0)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Activation,Flatten
from tensorflow.keras.optimizers import Adam
from sklearn import metrics
```

```python
# Number of Classes
num_labels=y_one_hot.shape[1]

# Creating an ann
model=Sequential()

# first layer
model.add(Dense(200,input_shape=(50,)))
model.add(Activation('relu'))
model.add(Dropout(0.5))

# second layer
model.add(Dense(100))
model.add(Activation('relu'))
model.add(Dropout(0.5))

# third layer
model.add(Dense(200))
model.add(Activation('relu'))
model.add(Dropout(0.5))

# final layer
model.add(Dense(num_labels))
model.add(Activation('softmax'))

model.summary()
```

```python
# Compile the model
model.compile(loss='categorical_crossentropy',metrics=['accuracy'],optimizer='adam')
```

```python
# Trianing the model
from tensorflow.keras.callbacks import ModelCheckpoint

num_epochs = 100
num_batch_size = 32

checkpointer = ModelCheckpoint(filepath='best_models/audio_classification.keras', 
                               verbose=1, save_best_only=True)

ann_history = model.fit(X_train, y_train, batch_size=num_batch_size, epochs=num_epochs, validation_data=(X_test, y_test), callbacks=[checkpointer], verbose=1)
```

```python
# Plotting accuracy
plt.figure(figsize=(8, 3))

plt.subplot(121)
plt.plot(ann_history.history['accuracy'], label='Training accuracy')
plt.plot(ann_history.history['val_accuracy'], label='val accuracy')
plt.title('Accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend()

# Plotting loss
plt.subplot(122)
plt.plot(ann_history.history['loss'], label='Training loss')
plt.plot(ann_history.history['val_loss'], label='val loss')
plt.title('Loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()
```

```python
from sklearn.metrics import accuracy_score
predict_ann = np.argmax(model.predict(X_test), axis=-1)

accuracy_ann = accuracy_score(y_test, np.array(pd.get_dummies(predict_ann))) * 100
print(f'Test Accuracy of ANN: {accuracy_ann} %')
```

## CNN

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y_one_hot,test_size=0.2,random_state=0)
```

```python
y_train.shape
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Number of Classes
num_classes=y_one_hot.shape[1]

# Input Shape
input_shape=(50,1)

# Define CNN model architecture
cnn_model = models.Sequential([
    layers.Conv1D(filters=32, kernel_size=3, activation='relu', input_shape=input_shape),
    layers.MaxPooling1D(pool_size=2),
    layers.Conv1D(filters=64, kernel_size=3, activation='relu'),
    layers.MaxPooling1D(pool_size=2),
    layers.Conv1D(filters=128, kernel_size=3, activation='relu'),
    layers.GlobalAveragePooling1D(),
    layers.Dense(64, activation='relu'),
    layers.Dense(num_classes, activation='softmax')
])

# Summary of the cnn model
cnn_model.summary()
```

```python
# Compile the model
cnn_model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
```

```python
# Train the model
cnn_history = cnn_model.fit(X_train, y_train, epochs=60, batch_size=32, validation_data=(X_test, y_test))
```

```python
# Plotting accuracy
plt.figure(figsize=(8, 3))

plt.subplot(121)
plt.plot(cnn_history.history['accuracy'], label='Training accuracy')
plt.plot(cnn_history.history['val_accuracy'], label='val accuracy')
plt.title('Accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend()

# Plotting loss
plt.subplot(122)
plt.plot(cnn_history.history['loss'], label='Training loss')
plt.plot(cnn_history.history['val_loss'], label='val loss')
plt.title('Loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()
```

```python
from sklearn.metrics import accuracy_score
# predicting results for test dataset
predict_cnn = np.argmax(cnn_model.predict(X_test), axis=-1)

accuracy_cnn = accuracy_score(y_test, np.array(pd.get_dummies(predict_cnn))) * 100
print(f'Test Accuracy of CNN: {accuracy_cnn} %')
```
